```python
승객수=pd.read_csv('../../Data/회귀분석/서울시 행정동별 대중교통 총 승차 승객수 정보.csv', encoding='ANSI')
승객수[승객수['기준_날짜']==20220101]
승객수= 승객수.astype({'기준_날짜':'str'})
승객수2022= 승객수[승객수['기준_날짜'].str.contains('2022')]
승객수2022
```

```python
mm= pd.merge(행정동, 승객수2022, on='행정동_ID')
mmm= mm.groupby(['자치구_명칭','행정동_명칭']).sum('승객_수')
```

```python
df=mmm[["승객_수"]]
df.reset_index(inplace=True)
```

```python
df.to_csv('../../Data/회귀분석/동별대중교통이용승객수.csv',encoding="utf-8")
```
