```python
df = pd.read_csv("data/서울시 건축물대장 정보.csv",encoding="ANSI", 
                usecols=["관리_건축물대장_PK","시군구_코드", "법정동_코드",
                        "번", "지","새주소_도로_코드", "새주소_본_번",
                         "새주소_부_번"])
df.info()
```

```python
df[["새주소_도로_코드","새주소_본_번","새주소_부_번"]]= df[["새주소_도로_코드","새주소_본_번","새주소_부_번"]].fillna(0)
df[["새주소_본_번","새주소_부_번"]]= df[["새주소_본_번","새주소_부_번"]].astype(int)
df.info()
```

```python
df[["번","지","새주소_도로_코드","새주소_본_번","새주소_부_번"]]= df[["번","지",
                "새주소_도로_코드","새주소_본_번","새주소_부_번"]].astype(str)
df.info()
```

```python
new_addr=[]

for i in range(len(df["관리_건축물대장_PK"])) :
    if df["새주소_도로_코드"][i]!="0" :
        if df["새주소_부_번"][i]!="0" :
            addr=df["새주소_도로_코드"][i]+" "+df["새주소_본_번"][i]+"-"+df["새주소_부_번"][i]
        else :
            addr=df["새주소_도로_코드"][i]+" "+df["새주소_본_번"][i]
    else :
        addr="NaN"
    
    if addr=="NaN" :
        new_addr.append(addr)
    else :
        new_addr.append(addr)
```

```python
old_addr=[]

for i in range(len(df["관리_건축물대장_PK"])) :
    if df["지"][i]!="0" :
        addr=df["시군구_코드"][i]+" "+df["법정동_코드"][i]+" "+df["번"][i]+"-"+df["지"][i]        
    else :
        addr=df["시군구_코드"][i]+" "+df["법정동_코드"][i]+" "+df["번"][i]
    old_addr.append("서울특별시 "+addr)
```
