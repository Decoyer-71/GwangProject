출처: https://data.seoul.go.kr/dataList/10406/S/2/datasetView.do

```python
import pandas as pd
import numpy as np
```

```python
file= pd.read_excel('../../Data/기타/생활안전도(2011_2018)_20230403172537.xlsx', header=[2,4], index_col=0)
file.columns.names=['구 별', '심각도']
file
```

```python
gu_list= ['종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구','은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구','강남구','송파구','강동구']
gu_list
```

```python
gu_prob=[]
for gu in gu_list: 
    gu_prob.append([gu,file[gu].sort_values(by='심각', ascending=False).iloc[0].name])
gu_prob

# 주차질서가 문제인 구가 너무 많음..
```
