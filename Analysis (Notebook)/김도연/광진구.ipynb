```python
import pandas as pd
from tqdm import trange

for tt in trange(10):
    driver = webdriver.Chrome("C:\\rlajs\pythonwork\\Analysis\\chromedriver")
    driver.get("https://parking.seoul.go.kr/")
    
    driver.find_element(By.CLASS_NAME, 'ico1').click()
    driver.find_element(By.CLASS_NAME, 'ico2').click()
    driver.find_element(By.CLASS_NAME, 'ico3').click()
    
    inpBox2= driver.find_element(By.ID, 'inpBox2')
    inpBox2.click()
    
    gus= inpBox2.find_element(By.CLASS_NAME, 'region_gu')
    gus.click()
    
    s_list=gus.find_element(By.CLASS_NAME,"select_list")
    time.sleep(1)
    gwang=s_list.find_elements(By.TAG_NAME,"a")[6]
    gwang.click()
    time.sleep(2)
    driver.find_element(By.ID,"parking_search").click()
    time.sleep(2)
    views= driver.find_element(By.ID, 'pk_search_view')
    time.sleep(2)
    lists= views.find_elements(By.TAG_NAME, 'ul')
#     views.find_element(By.TAG_NAME, 'a').click()
#     print(len(lists))
    cnt=1
    print(len(lists))
    for i in lists:
#         print(i.text)
    #     img= i.find_element(By.TAG_NAME, 'img')
    #     img.get_attribute('.*/ico1.jpg')
#         print(cnt)
        cnt+=1
        time.sleep(1)
        
    list=[i.text.split('\n') for i in lists]
    pd.DataFrame(list)
    
    name=[i[0] for i in list]
    park=[i[1] for i in list]
    df=pd.DataFrame({"주차장명":name, "정보":park})
    df1=df[df["정보"].str.contains("주차가능면")]
    total=[i[0] for i in df1["정보"].str.split("|")]
    park=[i[1] for i in df1["정보"].str.split("|")]
    df1["전체주차면수"]=total
    df1["주차가능면"]=park
    df1["전체주차면수"]=df1["전체주차면수"].str.extract("(\d+)").astype("float")
    df1["주차가능면"]=df1["주차가능면"].str.extract("(\d+)").astype("float")
    df1["주차비율"]=(df1["전체주차면수"]-df1["주차가능면"])/df1["전체주차면수"]*100
    
    df1.to_csv(f'{tt}.csv')
    print('end')
    driver.close()
    
    time.sleep(180)    
```
