```python
def get_dong_name(row):
    code = row['법정동코드']
    if code == 10100:
        return '중곡동'
    elif code == 10200:
        return '능동'
    elif code == 10300:
        return '구의동'
    elif code == 10400:
        return '광장동'
    elif code == 10500:
        return '자양동'
    elif code == 10700:
        return '화양동'
    elif code == 10800:
        return '모진동'
    elif code == 10900:
        return '군자동'
    else:
        return '-'

df['동명'] = df.apply(get_dong_name, axis=1)
```
