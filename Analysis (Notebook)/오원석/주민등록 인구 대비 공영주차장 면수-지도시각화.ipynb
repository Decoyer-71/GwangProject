## 구별 주민등록 인구

```python
people = pd.read_csv("C:/Users/acorn/Desktop/오원석/중간프로젝트/광진구/주민등록인구(월별)_20230405092746.csv")
people.shape
```

```python
people = people.drop([0,1,2]).reset_index(drop=True)
people
```

```python
people = people.drop("동별(1)", axis=1)
people
```

```python
people = people.rename(columns={'동별(2)': '구별', '2023. 02': '합계(명)', '2023. 02.1':'남자(명)',
                               '2023. 02.2':'여자(명)'})
people
```

```python
result = pd.merge(park, people, on='구별')
result.set_index('구별',inplace=True)
```

```python
result=result.astype(float)
result.info()
```

```python
result.sort_values('면수')
```

```python
result['인구대비 공영주차장 면수']=result['합계(명)']/result['면수']
result.sort_values('인구대비 공영주차장 면수')
```
