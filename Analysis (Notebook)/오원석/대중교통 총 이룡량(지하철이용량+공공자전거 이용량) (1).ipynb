```python
import numpy as np
import pandas as pd
```

```python
station=pd.read_csv('C:\\Users\\acorn\\Desktop\\오원석\\중간프로젝트\\광진구\\대중교통\\서울시 구별 지하철이용량(평균).csv')
station.round(0)
```

```python
bike=pd.read_csv('C:\\Users\\acorn\\Desktop\\오원석\\중간프로젝트\\광진구\\대중교통\\서울시 구별 공공자전거 이용량(평균).csv')
bike
```

```python
transportation=pd.concat([station,bike],axis=1)
transportation.drop('대여소 그룹',axis=1,inplace=True)
transportation.rename(columns={'승차총승객수': '지하철 승차총승객수','하차총승객수':'지하철 하차총승객수','총승객수':'지하철 총승객수',
                              '대여 건수':'공공자전거 대여 건수'}, inplace=True)
transportation
```

```python
transportation['대중교통 총이용량'] = transportation['지하철 총승객수'] + transportation['공공자전거 대여 건수']
```

```python
transportation.round()
```

```python
transportation.round().sort_values(by='대중교통 총이용량',ascending=False)
```
