```python
import os
import pandas as pd
import glob

file_path = r'C:\\Users\\acorn\\Desktop\\오원석\\중간프로젝트\\광진구\\주차장\\주차장 크롤링\\주차\\*.csv'

all_files = glob.glob(file_path)

# keys 옵션으로 파일명을 인덱스로 지정
df = pd.concat([pd.read_csv(f) for f in all_files], keys=[os.path.basename(f).replace('.csv', '') for f in all_files])
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df=df.reset_index()
df
```

```python
건국대학교병원=df[df['주차장명']=='건국대학교병원 주차장'][['주차비율','level_0']]
건국대학교사범대학부속중학교=df[df['주차장명']=='건국대학교사범대학부속중학교 주차장'][['주차비율','level_0']]
뚝섬1주차장=df[df['주차장명']=='뚝섬1주차장[윈드서핑장 앞]'][['주차비율','level_0']]
뚝섬2주차장=df[df['주차장명']=='뚝섬2주차장[안내센터 앞]'][['주차비율','level_0']]
뚝섬3주차장=df[df['주차장명']=='뚝섬3주차장[수영장 앞]'][['주차비율','level_0']]
뚝섬4주차장=df[df['주차장명']=='뚝섬4주차장[청담대교 하부]'][['주차비율','level_0']]
래미안프리미어팰리스주차장=df[df['주차장명']=='래미안프리미어팰리스 주차장'][['주차비율','level_0']]
보건복지행정타운=df[df['주차장명']=='보건복지행정타운'][['주차비율','level_0']]
성진강변스파랜드=df[df['주차장명']=='성진강변스파랜드'][['주차비율','level_0']]
스테이더디자이너스군자오피스텔주차장=df[df['주차장명']=='스테이더디자이너스군자오피스텔 주차장'][['주차비율','level_0']]
실로암사우나중곡점주차장=df[df['주차장명']=='실로암사우나 중곡점 주차장'][['주차비율','level_0']]
연한빌딩=df[df['주차장명']=='연한빌딩'][['주차비율','level_0']]
와이엠프라젠스파=df[df['주차장명']=='와이엠프라젠스파'][['주차비율','level_0']]
편안한요양병원=df[df['주차장명']=='편안한요양병원[광진자양점]'][['주차비율','level_0']]

merged_df = pd.merge(건국대학교병원, 건국대학교사범대학부속중학교,on='level_0', how='outer')\
.merge(뚝섬1주차장,on='level_0', how='outer').merge(뚝섬2주차장,on='level_0', how='outer')\
.merge(뚝섬3주차장,on='level_0', how='outer').merge(뚝섬4주차장,on='level_0', how='outer')\
.merge(래미안프리미어팰리스주차장,on='level_0', how='outer').merge(보건복지행정타운,on='level_0', how='outer')\
.merge(성진강변스파랜드,on='level_0', how='outer').merge(스테이더디자이너스군자오피스텔주차장,on='level_0', how='outer')\
.merge(실로암사우나중곡점주차장,on='level_0', how='outer').merge(연한빌딩,on='level_0', how='outer')\
.merge(와이엠프라젠스파,on='level_0', how='outer').merge(편안한요양병원,on='level_0', how='outer')
# merged_df=merged_df.set_index('level_0')
merged_df

merged_df = pd.DataFrame(merged_df.values, columns=['건국대학교병원','날짜','건국대학교사범대학부속중학교','뚝섬1주차장','뚝섬2주차장',
                                                   '뚝섬3주차장','뚝섬4주차장','래미안프리미어팰리스주차장','보건복지행정타운',
                                                   '성진강변스파랜드','스테이더디자이너스군자오피스텔주차장','실로암사우나중곡점주차장',
                                                   '연한빌딩','와이엠프라젠스파','편안한요양병원'])
merged_df.set_index('날짜',inplace=True)
merged_df.to_csv('부설주차장 이용량.csv')
```
