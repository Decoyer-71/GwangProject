```python
prob = pd.read_excel("./data/생활안전도(2011_2018)_20230404164731.xlsx", header = 4, index_col= 0, 
                     names = ["단독주택", "아파트", "다세대주택", "연립주택/기타"])
prob
```

```python
# 광진구 내 주택(공동주택, 단독주택)별 생활안전도 조사 결과 시각화
label = ["주차", "쓰레기 방치", "대기오염", "휴식공간/\n녹지부족", 
         "소음", "범죄", "수질오염"]
plt.pie(prob["단독주택"].values, labels= label,
       startangle = 270,
        autopct = "%.2f%%",
       counterclock = True,
        labeldistance = 1.1,
        explode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05],
        shadow = True,
        colors = ["lightgreen", "lightgray", "lightgray", "lightgray", 
                  "lightgray", "lightgray", "lightgray"],
       wedgeprops = {"width" : 0.7, "linewidth" : 3},
       textprops = {"size" : 12})
plt.title("생활안전도 조사결과 [단독주택]", fontsize = 15, loc = "center")
plt.tight_layout()
```
