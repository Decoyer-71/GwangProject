```python
import pandas as pd
```

```python
df = pd.read_csv('/content/서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv',encoding='cp949')
df.head()
```

```python
times = []
for i in range(24):
  a = str(i).zfill(2)
  b = str(i+1).zfill(2)
  times.append('{}시-{}시'.format(a,b))
```

```python
for time in times:
  df[time + ' 승하차 인원'] = df[time + ' 승차인원']+df[time + ' 하차인원']
```

```python
df2=df.drop(df.columns[3:52],axis=1)
df2
```

```python
df3=df2.drop('사용월',axis=1).groupby(['호선명','지하철역']).mean().round(0)
df3.head()
```
