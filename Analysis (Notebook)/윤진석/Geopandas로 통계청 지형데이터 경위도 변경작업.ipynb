```python
a = data[data["EMD_NM"] == "광장동"]["geometry"]
a
```

```python
b = str(a.values)
```

```python
import re
dong_list = re.findall("(\d+)", b)
dong_list
```

```python
s1 = []
s2 = []
for i in range(len(dong_list)):
    if i % 2 == 0:
        s1.append(dong_list[i])
    elif i % 2 == 1:
        s2.append(dong_list[i])
print(s1)
print(s2)
```

```python
ssibal1 = []
for i in range(len(s1)):
    for j in range(len(s2)):
        if i == j:
            ssibal1.append(s1[i] + "." + s2[j])
print(ssibal1)
```

```python
ssibal2 = []
try : 
    for i in range(len(ssibal1)):
        ssibal2.append([ssibal1[2*i] +", "+ ssibal1[2*i+1]])
except : None
print(ssibal2)
```

```python
import re
ssibal3 = []
for i in ssibal2:
    ssibal3.append(re.sub("'", "", str(i)))
```

```python
for i in ssibal3:
    print(i+",")
```
